## ANÁLISE DOS PERFIS DE CAMPO DO ONDULADOR DELTA SABIÁ CONSIDERANDO FORMA PADRÃO DO BLOCO MAS COM DIMENSÕES CORRIGIDAS 
**GV = -PERIODO/2**

```python
from imaids import models 
import numpy as np 
import matplotlib.pyplot as plt 

delta52_1 = models.DeltaSabia()
delta52_1.dgv = -delta52_1.period_length/2
delta52_1.solve()

delta52_2 = models.DeltaSabia(block_shape = shape)
delta52_2.dgv = -delta52_2.period_length/2
delta52_2.solve()
```

```python
np.array(delta52_1.calc_field_amplitude()) - np.array(delta52_2.calc_field_amplitude()) 
```

```python
z_list = np.linspace(-1000,1000,2000)
field_1 = delta52_1.get_field(z = z_list)
field_2 = delta52_2.get_field(z = z_list)
```

```python
figure, axes = plt.subplots(nrows = 2, ncols = 3, figsize=(25,10))
axes[0,0].plot(z_list, field_1[:,0])
axes[0,0].plot(z_list, field_2[:,0], '--')
axes[0,1].plot(z_list, field_1[:,1])
axes[0,1].plot(z_list, field_2[:,1],'--')
axes[0,2].plot(z_list, field_1[:,2], label = 'Default Shape')
axes[0,2].plot(z_list, field_2[:,2], '--', label = 'Fixed Shape')
axes[0,2].legend()
axes[0,0].set(ylabel = 'Bx [T]')
axes[0,1].set(ylabel = 'By [T]')
axes[0,2].set(ylabel = 'Bz [T]')

axes[1,0].plot(z_list, np.array(field_1[:,0]) - np.array(field_2[:,0]))
axes[1,1].plot(z_list, np.array(field_1[:,1]) - np.array(field_2[:,1]))
axes[1,2].plot(z_list, np.array(field_1[:,2]) - np.array(field_2[:,2]), label = "Default - Fixed")

axes[1,2].legend()
axes[1,0].set(ylabel = 'Bx [T]',xlabel = 'z [mm]')
axes[1,1].set(ylabel = 'By [T]',xlabel = 'z [mm]')
axes[1,2].set(ylabel = 'Bz [T]',xlabel = 'z [mm]')

axes[0,1].set_title('Magnetic Field for Delta Sabia Undulator')
axes[1,1].set_title('Magnetic Field Difference for different Block Shape')

axes[0,0].grid(True)
axes[0,1].grid(True)
axes[0,2].grid(True)
axes[1,0].grid(True)
axes[1,1].grid(True)
axes[1,2].grid(True)

plt.show()
```

```python
bx_diff_porcent = max(np.array(field_1[:,0]) - np.array(field_2[:,0]))/max(np.array(field_2[:,0]))*100
by_diff_porcent = max(np.array(field_1[:,1]) - np.array(field_2[:,1]))/max(np.array(field_2[:,1]))*100
bz_diff_porcent = max(np.array(field_1[:,2]) - np.array(field_2[:,2]))/max(np.array(field_2[:,2]))*100

print("Bx Diff Porcentage = {:.4f}%".format(bx_diff_porcent))
print("By Diff Porcentage = {:.4f}%".format(by_diff_porcent))
print("Bz Diff Porcentage = {:.4f}%".format(bz_diff_porcent))
```

```python
x_list = np.linspace(-5,5,11)

ibx_1 = []
iby_1 = []

ibx_2 = []
iby_2 = []

for m in x_list:
    ib1, iib1 = delta52_1.calc_field_integrals(z_list = z_list, x = m, y = 0)
    ibx_1.append(ib1[-1,0])
    iby_1.append(ib1[-1,1])
    
    ib2, iib2 = delta52_2.calc_field_integrals(z_list = z_list, x = m, y = 0)
    ibx_2.append(ib2[-1,0])
    iby_2.append(ib2[-1,1])
```

```python
figure, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(25,10))
axes[0,0].plot(x_list, ibx_1)
axes[0,0].plot(x_list, ibx_2, '--')
axes[0,1].plot(x_list, iby_1, label = 'Default Shape')
axes[0,1].plot(x_list, iby_2, '--', label = 'Fixed Shape')
axes[0,1].legend()
axes[0,0].set(ylabel = 'IBx [G.cm]')
axes[0,1].set(ylabel = 'IBy [G.cm]')

axes[1,0].plot(x_list, np.array(ibx_1) - np.array(ibx_2))
axes[1,1].plot(x_list, np.array(iby_1) - np.array(iby_2))

axes[1,0].set(ylabel = 'IBx [G.cm]',xlabel = 'x [mm]')
axes[1,1].set(ylabel = 'IBy [G.cm]',xlabel = 'x [mm]')


axes[0,1].set_title('Field Integral for Delta Sabia Undulator')
axes[1,1].set_title('Field Integral Difference for different Block Shape')

axes[0,0].grid(True)
axes[0,1].grid(True)
axes[1,0].grid(True)
axes[1,1].grid(True)

plt.show()
```
